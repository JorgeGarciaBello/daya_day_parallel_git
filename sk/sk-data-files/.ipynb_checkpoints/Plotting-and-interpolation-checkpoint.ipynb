```python
# Cargando paquetes necesarios
import pandas as pd
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
# Cargar base de datos: (3D-parameters, 1-chi2)
# Super-Kamiokande Data
data1 = pd.read_csv('sk-nh-ji.dat',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)
#data1 = pd.read_csv('sk-ih-ji.dat',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)

# RENO Data
data2 = pd.read_csv('reno-3D-nh.dat',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)
#data2 = pd.read_csv('reno-3D-ih.dat',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)

# Daya Bay Data
data3 = pd.read_csv('db-3D-nh.dat',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)
#data3 = pd.read_csv('db-3D-ih.dat',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)

# MINOS data
data4 = pd.read_csv('minos.3D.grid.of.sk.dm32.s2t13.s2t23.nh',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)
#data4 = pd.read_csv('minos.3D.grid.of.sk.dm32.s2t13.s2t23.ih',names=['dm32','s2_t13','s2_t23','ji'],delim_whitespace=True)

all_data=[data1,data2,data3,data4]
```

### Normalizando datos a ji=0

```python
##########################################################################################################
#
#  get_min_vals: Is is a function that return an array with one row per data set,
#                and columns [idx_min,dm32_min,s2_t13_min,s2_t23_min,ji_min].
#                Where:
#                 idx_min - is the index with the value of ji's minimum
#                 ji_min  - is the ji's minimum
#                 dm32_min  -  is the minimum value of the parameter dm32  
#                 s2_t13_min  -  is the minimum value of the parameter s2_t13
#                 s2_t23_min  -  is the minimum value of the parameter s2_t23
#
###########################################################################################################
def get_min_vals(data_array):
    result=[]
    for i in range(0,len(data_array)):        
        idx_min    = data_array[i]['ji'].idxmin()
        ji_min     = data_array[i]['ji'][idx_min]
        dm32_min   = data_array[i]['dm32'][idx_min]
        s2_t13_min = data_array[i]['s2_t13'][idx_min]
        s2_t23_min = data_array[i]['s2_t23'][idx_min]
        r=(dm32_min,s2_t13_min,s2_t23_min,ji_min)
        result.append(r)
    return result
```

```python
# Obteniendo los valores minimos por parámetro en cada conjunto de datos
dat_mins=get_min_vals(all_data)
dat_mins
```

```python
def get_normalize_data(data_array):
    result=[]
    dat_mins=get_min_vals(data_array)
    for i in range(0,len(data_array)):        
        data_array[i]['ji']  = data_array[i]['ji'] - dat_mins[i][3]
        result.append(data_array[i])
    return result    
```

```python
data_norm = get_normalize_data(all_data)
```

### Filtrar datos pra generar perfiles de confianza - (parameter, Ji)

```python
def get_data_to_plotter_1D(data_array):
    result=[]    
    for e in range(0,len(data_array)):
        data = data_array[e]
        dm32_vs_ji   = data[(data['s2_t13'] == dat_mins[e][1]) & (data['s2_t23'] == dat_mins[e][2])][['dm32','ji']]
        s2_t13_vs_ji = data[(data['dm32']   == dat_mins[e][0]) & (data['s2_t23'] == dat_mins[e][2])][['s2_t13','ji']]
        s2_t23_vs_ji = data[(data['dm32']   == dat_mins[e][0]) & (data['s2_t13'] == dat_mins[e][1])][['s2_t23','ji']]
        result.append([dm32_vs_ji,s2_t13_vs_ji,s2_t23_vs_ji])
    return result
```

```python
data_to_plotter_1D = get_data_to_plotter_1D(data_norm)
```

### Filtrar datos pra generar regiones de confianza - (parameter1,parameter2, Ji)

```python
def get_data_to_plotter_2D(data_array):
    result=[]
    for e in range(0,len(data_array)):
        data = data_array[e]
        dm32_vs_s2_t13_ji   = data[(data['s2_t23'] == dat_mins[e][2])][['dm32',  's2_t13', 'ji']]
        dm32_vs_s2_t23_ji   = data[(data['s2_t13'] == dat_mins[e][1])][['dm32',  's2_t23', 'ji']]        
        s2_t13_vs_s2_t23_ji = data[(data['dm32']   == dat_mins[e][0])][['s2_t13','s2_t23', 'ji']]        
        result.append([dm32_vs_s2_t13_ji,dm32_vs_s2_t23_ji,s2_t13_vs_s2_t23_ji])
    return result
```

```python
data_to_plotter_2D = get_data_to_plotter_2D(data_norm)
```

### Interpolación de datos antes de graficar

```python
##########################################################################################################
#
#  get_interpolated_data_1D: It is a function that return an array with interpolated data.
#        Input: data_array - It is an array from 'get_data_to_plotter_1D' function. 
#        
#        Output: [array] - Each row containing three DataFrames (DF1,DF2,DF3), with the next structure:
#
#               ROW => [DF1,DF2,DF3], where DF1 => ['dm32',  'ji'] 
#                                           DF2 => ['s2_t13','ji']
#                                           DF3 => ['s2_t23','ji']
#
###########################################################################################################
def get_interpolated_data_1D(data_array):
    result=[]
    charts=['dm32','s2_t13','s2_t23']
    for e in range(0,len(data_array)):
        chart_results=[]
        for chart in range(0,len(charts)):
            x=data_array[e][chart][charts[chart]].values
            y=data_array[e][chart].ji.values
            f1 = interpolate.interp1d(x, y, kind = 'cubic')
            x_new = np.linspace(x[0], x[-1],100)
            y_new = f1(x_new)
            data_new = pd.DataFrame({charts[chart]:x_new,'ji':y_new})
            chart_results.append(data_new)
        result.append(chart_results)
    return result
```

```python
def get_interpolated_data_2D(data_array):
    result=[]
    size=1000
    charts=[['dm32','s2_t13'],['dm32','s2_t23'],['s2_t13','s2_t23']]
    for e in range(0,len(data_array)):
        chart_results=[]
        for chart in range(0,len(charts)):
            x=data_array[e][chart][charts[chart][0]].drop_duplicates().to_numpy()
            y=data_array[e][chart][charts[chart][1]].drop_duplicates().to_numpy()
            z=data_array[e][chart].ji.values            
            f = interpolate.interp2d(x, y, z, kind='cubic')
            x_new=np.linspace(x[0], x[-1],size)
            y_new=np.linspace(y[0], y[-1],size)
            z_new=f(x_new,y_new).reshape(size*size)
            x=[]
            y=[]
            for i in x_new:
                for j in y_new:
                    x.append(i)
                    y.append(j)            
            data_new = pd.DataFrame({charts[chart][0]:x,charts[chart][1]:y,'ji':z_new})
            chart_results.append(data_new)
        result.append(chart_results)
    return result
```

```python
interpolated_data_1D = get_interpolated_data_1D(data_to_plotter_1D)
interpolated_data_2D = get_interpolated_data_2D(data_to_plotter_2D)
```

### Creando gráficas ( Parameter vs Ji )

```python
def plot_dm32_vs_ji(obj,label_n):
    #fig, ax = plt.subplots(figsize=(9,6))
    fig, ax = plt.subplots()
    
    ax.plot('dm32', 'ji', data=obj,label=label_n,c='black')#,linestyle='-.')#,label=r'Results from $\chi^2_{RENO}$')
    lgnd = plt.legend(loc="lower right",fontsize=13)#, scatterpoints=1, )
    ax.set_xlim(0.001, 0.005)
    ax.set_ylim(0, 20)    

    ax.set_xticks([0.001,0.002,0.003,0.004,0.005])
    ax.set_yticks([0,5,10,15,20])
    
    ax.set_xticklabels(['$0.001$','$0.002$','$0.003$','$0.004$','$0.005$'])
    ax.set_yticklabels(['$0$','$5$','$10$','$15$','$20$'])    
    
    ax.tick_params(labelcolor='black', labelsize=13, width=1)
    ax.set(xlabel=r'$\Delta m^2_{32} \hspace{0.5} [eV^{2}] \hspace{0.5} (x10^{-3})$', ylabel=r'$\Delta \chi^2 $')
    ax.xaxis.label.set_size(14)
    ax.yaxis.label.set_size(14)
    plt.tight_layout()
    ax.grid(True, linestyle='--')
    fig.savefig('plot_'+label_n+'_dm32_vs_ji.png')    
    return 

def plot_s2_t13_vs_ji(obj,label_n):
    #fig, ax = plt.subplots(figsize=(9,6))
    fig, ax = plt.subplots()
    
    ax.plot('s2_t13', 'ji', data=obj,label=label_n,c='black')#,linestyle='-.')#,label=r'Results from $\chi^2_{RENO}$')
    lgnd = plt.legend(loc="lower right",fontsize=13)#, scatterpoints=1, )
    
    ax.set_xlim(0.00, 0.065)
    ax.set_ylim(0, 20)

    ax.set_xticks([0.0,0.02,0.04,0.06])
    ax.set_yticks([0,5,10,15,20])
    
    ax.set_xticklabels(['$0.0$','$0.02$','$0.04$','$0.06$'])
    ax.set_yticklabels(['$0$','$5$','$10$','$15$','$20$'])    
    
    ax.tick_params(labelcolor='black', labelsize=13, width=1)
    
    ax.set(xlabel=r'$\sin^2 \theta_{13}$', ylabel=r'$\Delta \chi^2 $')
    ax.xaxis.label.set_size(14)
    ax.yaxis.label.set_size(14)
    plt.tight_layout()
    ax.grid(True, linestyle='--')
    fig.savefig('plot_'+label_n+'_s2_t13_vs_ji.png')        
    return 

def plot_s2_t23_vs_ji(obj,label_n):
    #fig, ax = plt.subplots(figsize=(9,6))
    fig, ax = plt.subplots()
    
    ax.plot('s2_t23', 'ji', data=obj,label=label_n,c='black')#,linestyle='-.')#,label=r'Results from $\chi^2_{RENO}$')
    lgnd = plt.legend(loc="lower right",fontsize=13)#, scatterpoints=1, )
    
    ax.set_xlim(0.2, 0.8)
    ax.set_ylim(0, 20)

    ax.set_xticks([0.2,0.4,0.6,0.8])
    ax.set_yticks([0,5,10,15,20])
    
    ax.set_xticklabels(['$0.2$','$0.4$','$0.6$','$0.8$'])
    ax.set_yticklabels(['$0$','$5$','$10$','$15$','$20$'])    
    
    ax.tick_params(labelcolor='black', labelsize=13, width=1)
    
    ax.set(xlabel=r'$\sin^2 \theta_{23}$', ylabel=r'$\Delta \chi^2 $')
    ax.xaxis.label.set_size(14)
    ax.yaxis.label.set_size(14)
    plt.tight_layout()
    ax.grid(True, linestyle='--')
    fig.savefig('plot_'+label_n+'_s2_t23_vs_ji.png')        
    return
def plot_parameters_vs_ji(obj,label_n):
    for i in range(0,len(obj)):
        plot_dm32_vs_ji(obj[i][0],label_n[i])
        plot_s2_t13_vs_ji(obj[i][1],label_n[i])
        plot_s2_t23_vs_ji(obj[i][2],label_n[i])
    return    
```

```python
plot_parameters_vs_ji(interpolated_data_1D,['sk','reno','db','minos'])
```
